```python
import alipy
from alipy import experiment
from alipy.experiment import ExperimentAnalyser
import os


def load_models_folds(model, num_files,strategy,directory=None):
    model_folds=[]
    if directory is None:
        directory='../../AL_Comparision/'+str(model)+'/'+str(model)+'_'+str(strategy)
    for i in range(num_files):
        file_name = 'AL_round_' + str(i) + '.pkl'
        file_path = os.path.join(directory, file_name)

        if os.path.isfile(file_path):
            # 这里可以对找到的文件进行操作或处理
            file = alipy.experiment.StateIO.load(file_path)
            model_folds.append(file)
            print('找到文件:', file_path)
        else:
            print('未找到文件:', file_path)
    return model_folds

strategy='GD'

#加载模型folds
LR_folds =load_models_folds(model='LR', num_files=10,strategy=strategy)
Ada_folds =load_models_folds(model='Ada', num_files=10,strategy=strategy)
DTC_Entropy_folds =load_models_folds(model='DTC_Entropy', num_files=10,strategy=strategy,
                                     directory='../../AL_Comparision/DTC_Entropy/DTC_GD')
DTC_Gini_folds=load_models_folds(model='DTC_Gini', num_files=10,strategy=strategy,
                                     directory='../../AL_Comparision/DTC_Gini/DTC_GD')
GP_folds=load_models_folds(model='GP', num_files=10,strategy=strategy)
KNN_folds =load_models_folds(model='KNN', num_files=10,strategy=strategy)
LDA_folds =load_models_folds(model='LDA', num_files=10,strategy=strategy)
MLP_folds =load_models_folds(model='MLP', num_files=10,strategy=strategy)
NB_folds=load_models_folds(model='NB', num_files=10,strategy=strategy)
QDA_folds=load_models_folds(model='QDA', num_files=1,strategy=strategy)
RBF_folds=load_models_folds(model='RBF', num_files=10,strategy=strategy)
RFC_folds=load_models_folds(model='RFC', num_files=10,strategy=strategy)
SGD_folds=load_models_folds(model='SGD', num_files=10,strategy=strategy)
SVC_folds=load_models_folds(model='SVC', num_files=10,strategy=strategy)
#初始化StateIOContainer
StateIOContainer_LR = alipy.experiment.StateIOContainer(method_name='LR', method_results=[])
StateIOContainer_Ada = alipy.experiment.StateIOContainer(method_name='Ada', method_results=[])
StateIOContainer_DTC_Entropy = alipy.experiment.StateIOContainer(method_name='DTC_Entropy', method_results=[])
StateIOContainer_DTC_Gini = alipy.experiment.StateIOContainer(method_name='DTC_Gini', method_results=[])
StateIOContainer_GP = alipy.experiment.StateIOContainer(method_name='GP', method_results=[])
StateIOContainer_KNN = alipy.experiment.StateIOContainer(method_name='KNN', method_results=[])
StateIOContainer_LDA = alipy.experiment.StateIOContainer(method_name='LDA', method_results=[])
StateIOContainer_MLP = alipy.experiment.StateIOContainer(method_name='MLP', method_results=[])
StateIOContainer_NB = alipy.experiment.StateIOContainer(method_name='NB', method_results=[])
StateIOContainer_QDA = alipy.experiment.StateIOContainer(method_name='QDA', method_results=[])
StateIOContainer_RBF = alipy.experiment.StateIOContainer(method_name='RBF', method_results=[])
StateIOContainer_RFC = alipy.experiment.StateIOContainer(method_name='RFC', method_results=[])
StateIOContainer_SGD = alipy.experiment.StateIOContainer(method_name='SGD', method_results=[])
StateIOContainer_SVC = alipy.experiment.StateIOContainer(method_name='SVC', method_results=[])
#将folds装载进StateIOContainer
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_LR, folds=LR_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_Ada, folds=Ada_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_DTC_Entropy, folds=DTC_Entropy_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_DTC_Gini, folds=DTC_Gini_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_GP, folds=GP_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_KNN, folds=KNN_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_LDA, folds=LDA_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_MLP, folds=MLP_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_NB, folds=NB_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_QDA, folds=QDA_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_RBF, folds=RBF_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_RFC, folds=RFC_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_SGD, folds=SGD_folds)
alipy.experiment.StateIOContainer.add_folds(self=StateIOContainer_SVC, folds=SVC_folds)
#使用analyser分析绘图
analyser1 = ExperimentAnalyser()
analyser2 = ExperimentAnalyser()
analyser3 = ExperimentAnalyser()

analyser1.add_method(method_name='Ada', method_results=StateIOContainer_Ada)
analyser1.add_method(method_name='KNN', method_results=StateIOContainer_KNN)
analyser1.add_method(method_name='MLP', method_results=StateIOContainer_MLP)
analyser1.add_method(method_name='RFC', method_results=StateIOContainer_RFC)


analyser2.add_method(method_name='LDA', method_results=StateIOContainer_LDA)
analyser2.add_method(method_name='QDA', method_results=StateIOContainer_QDA)
analyser2.add_method(method_name='RBF', method_results=StateIOContainer_RBF)
analyser2.add_method(method_name='SGD', method_results=StateIOContainer_SGD)
analyser2.add_method(method_name='SVC', method_results=StateIOContainer_SVC)

analyser3.add_method(method_name='DTC_Entropy', method_results=StateIOContainer_DTC_Entropy)
analyser3.add_method(method_name='DTC_Gini', method_results=StateIOContainer_DTC_Gini)
analyser3.add_method(method_name='GP', method_results=StateIOContainer_GP)
analyser3.add_method(method_name='LR', method_results=StateIOContainer_LR)
analyser3.add_method(method_name='NB', method_results=StateIOContainer_NB)
#RFC、MLP、Ada、KNN
analyser1.plot_learning_curves(title='Comparison of QBC Strategies for Multiple Models')
#SVM（RBF）、SVM（SGD）、GP、LDA、QDA
analyser2.plot_learning_curves(title='Comparison of QBC Strategies for Multiple Models')
#LR、NB、DTC
analyser3.plot_learning_curves(title='Comparison of QBC Strategies for Multiple Models')
```
